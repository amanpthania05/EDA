```python
#OUTLIERS DETECTION FOR ALL COLUMNS
data_columns=['LB','AC','FM','UC','DL','DS','DP','ASTV','MSTV','ALTV','MLTV']
for i in data_columns:
  data_columns=i
  print("Outliers Calculation for" , i)
  print(data.boxplot(column=[data_columns],grid=True))
#QUARTILE 1
  print(data[data_columns].describe())
  Q1=data[data_columns].quantile(0.25)
  print('Quartile 1 :',Q1)
#QUARTILE 3
  Q3=data[data_columns].quantile(0.75)
  print('Quartile 3 :',Q3)
#INTERQUARTILE RANGE
  IQR=Q3-Q1
  print('Inter-Quartile Range :' ,IQR)
#UPPER LIMIT
  UL=Q3+(1.5*IQR)
  print('Upper Limit:',UL)
#LOWER LIMIT
  LL=Q3+(1.5*IQR)
  print('Lower Limit:',LL)
  Outliers=data[(data[data_columns] > UL) | (data[data_columns] < LL)]
  print(Outliers)
  print('_________________________________________________________________________')
```

# **MISSING VALUES AND IMPUTATION**

```python
data.isnull().sum()  #TO CHECK HOW MANY NULL VALUES ARE EXISTING IN DATA
```

```python
# FINDING % OF MISSING VALUE
print('the % of missing valiue in LB is',(data['LB'].isnull().sum()/len(data))*100)
print('the % of missing valiue in AC is',(data['AC'].isnull().sum()/len(data))*100)
print('the % of missing valiue in DS is',(data['DS'].isnull().sum()/len(data))*100)
print('the % of missing valiue in DP is',(data['DP'].isnull().sum()/len(data))*100)
print('the % of missing valiue in MLTV is',(data['MLTV'].isnull().sum()/len(data))*100)
```

If the dataset is small, and only 3-5% data is missing from a specific numerical column, then it will be ideal to replace the missing values with either mean or median

Mean: If the data is normally distributed
Median: If the data is skewed
If more than 60-70% is missing, then it will be ideal to drop the column, because imputing so much data is neither possible, nor appropriate.

If data is missing upto 5% to 40%, then used advanced techniques. Either use linear regression, decision trees, or discretization
